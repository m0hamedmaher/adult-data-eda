```python
mean = df['fnlwgt'].mean()
std_dev = df['fnlwgt'].std()

# Define threshold for outliers
threshold = 2  # Adjust as needed

# Identify outliers
df['is_outlier'] = abs(df['fnlwgt'] - mean) > threshold * std_dev

# Replace outliers with mean value
df.loc[df['is_outlier'], 'fnlwgt'] = mean
sns.boxplot(df['fnlwgt'])
```
